```python
import pandas as pd
import folium
```

```python
# Importing the file from github. Used to read the csv file using the raw data link. Placed into the dallas_zips object
url = 'https://raw.githubusercontent.com/AliceLiu17/KPMG-Sustainable-Future/main/Data/Zip%20Code%20(cleaned).csv'
dallas_zips= pd.read_csv(url)
dallas_zips.head(3)
```

```python
#Creating a map and giving it a name by using a map class. This will display a world map.
#Inside the map we can zoom in by giving it a pinpoint so we can zoom into the map. 
map = folium.Map(
    location =  [32.7767, -96.7970], #[latitude, longitude of Dallas]
    zoom_start = 10.5
)
```

```python
from folium.vector_layers import Popup
#Pinpoint the areas within the map using a marker.
#Uses a for loop to go through the csv file(dallas_zips) so it can iterate throughout the entire map(df.iterrows), espcialy, when we want to add
#We will be using folium.Marker()
print(dallas_zips.columns.tolist()) #Tells me columns are in the dallas_zips

for _, dallas in dallas_zips.iterrows():
  html = '''
    <b>Zip: </b> {status} <br>
    <b>Region: </b></b>{type} <br>
    <b>Average Income: </b></b>{contact} <br>
    <b>Median Age: </b></b>{title} <br>

    '''.format(status=dallas['Zip Code'],type=dallas['Region'],contact=dallas['Average Income Per Household  ($)'], title =dallas['Median Age (years)'] )


  folium.Marker(
      location=[dallas['Latitude '], dallas['Longitude']],
      popup=folium.Popup(html, parse_html=False, max_width=1000), #Adjusts the size of the popup
      tooltip=html # Allows you to hover over each icon without clicking
  ).add_to(map)

map
```

```python
# clean the data
dallas_zips = dallas_zips.loc[:, dallas_zips.columns.intersection(["Zip Code", "Latitude", "Longitude", "Current Population", "Average House Value ($)", "Average Income Per Household  ($)", "White Population", "Black Population", "Hispanic Population",	"Asian Population", "American Indian Population",	"Hawaiian Population",	"Other Population",	"Male Population",	"Female Population", "Median Age (years)",	"Male Median Age (years)",	"Female Median Age (years)"])]
dallas_zips.head()
```

```python
dallas_zips["Average Income Per Household  ($)"].isnull().sum()
```

```python
dallas_zips["Average Income Per Household  ($)"] = dallas_zips["Average Income Per Household  ($)"].replace(",", "", regex = True)
dallas_zips["Average Income Per Household  ($)"] = dallas_zips["Average Income Per Household  ($)"].astype(int)
dallas_zips["Average Income Per Household  ($)"].dtypes
```

```python
folium.Choropleth(
    geo_data = url,
    name = "choropleth",
    data = dallas_zips,
    columns = ["Zip Code", "Average Income Per Household  ($)"],
    key_on = "feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name = "Average Income Per Household"
).add_to(map)

folium.LayerControl().add_to(map)
map
```
