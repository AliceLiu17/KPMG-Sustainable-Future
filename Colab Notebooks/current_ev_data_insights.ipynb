```python
import numpy as np
import pandas as pd
from google.colab import files
```

```python
df = pd.read_csv("ev_stations.csv")
```

```python
dallas_df = df[df['City'] == 'Dallas']
dallas_df = dallas_df.reset_index()
dallas_df = dallas_df.drop(columns = ['index'])
dallas_df = dallas_df.loc[:, ~dallas_df.isnull().all()]
```

```python
dallas_zip_df = dallas_df[["ZIP", "Open Date"]]
dallas_zip_df['Open Date'] = dallas_zip_df['Open Date'].str[:4]
```

```python
d = dallas_zip_df.groupby(["ZIP", "Open Date"])['ZIP'].count().reset_index(name='Counts')
df = pd.DataFrame(data=d)
df = df.rename(columns={"ZIP": "zip_code", "Open Date": "year", "Counts" : 'count'})
```

```python
df.to_csv('current_evcs.csv') 
files.download('current_evcs.csv')
```
