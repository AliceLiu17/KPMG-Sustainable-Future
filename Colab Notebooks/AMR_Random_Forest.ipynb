```python
# created training and testing data sets
y = df["ev_registration_count"]
X = df.drop(columns = ["ev_registration_count", "evcs_count"], axis = 1)

# split data into training and testing data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 50)

# create the model
model = RandomForestRegressor(n_estimators = 31, max_features = 'sqrt', max_depth = 4, min_samples_split = 2, min_samples_leaf = 2, bootstrap = True, random_state = 48)

# fit the model to the training data
model.fit(X_train, y_train)

# make prediction on the test data
y_pred = model.predict(X_test)

# call the mean_squared_error() with y_test and prediction
model_rmse = mean_squared_error(y_test, y_pred, squared = False)

# r2 score:
rmse_r2 = r2_score(y_test, y_pred)

print("RMSE: {0}".format(model_rmse))
print("R2 Score: {0}".format(rmse_r2))
```

```python
parameters = {
    'n_estimators': [25, 50, 100, 150, 200, 250],
    'max_depth': [1,2,3,4],
}
regr = RandomForestRegressor(random_state=0)

clf = GridSearchCV(regr, parameters)
clf.fit(X_train, y_train)
y_pred_train = clf.predict(X_train)
mean_squared_error(y_train, y_pred_train)
print(clf.best_score_)
print(clf.best_params_)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 5, stop = 50, num = 40)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(2, 10, num = 8)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
rf_random.best_params_
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy

base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)

best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)

print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```
