```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import seaborn as sns
```

```python
rachel_final_url = "https://raw.githubusercontent.com/AliceLiu17/KPMG-Sustainable-Future/main/Data/df_final.csv"
df_final = pd.read_csv(rachel_final_url)

dallas_demo_update_url = "https://raw.githubusercontent.com/AliceLiu17/KPMG-Sustainable-Future/main/Data/Dataframe%20to%20combine/Dallas_demographics_readyformodel.csv"
df_demo = pd.read_csv(dallas_demo_update_url)

df_final.head(5)
```

```python
df_demo = df_demo.drop("Unnamed: 0", axis = 1)
```

```python
df_demo.columns
```

```python
a = df_demo["White Population"]
b = df_demo["Hispanic Population"]
c = df_demo["Black Population"]
d = df_demo["Asian Population"]
e = df_demo["American Indian Population"]
f = df_demo["Hawaiian Population"]
g = df_demo["Other Population"]
h = df_demo["Male Population"]
i = df_demo["Female Population"]
j = df_demo["# Residential Mailboxes"]
k = df_demo["# Business Mailboxes"]
l = df_demo["Number of Businesses"]

df_final = pd.concat([df_final, a, b, c, d, e, f, g, h, i, j, k, l], axis = 1)
df_final.columns
```

```python
df_final =  df_final.rename(columns={"EV Charging Station Count": "evcs_count",
                         "Current Population": "population",
                         "Average House Value ($)": "house_value",
                         "Average Income Per Household  ($)": "household_income",
                         "# Residential/Business Mailbox Ratio": "residential_business_ratio",
                         "EV Registration Count": "ev_registration_count"})
```

```python
x = df_final.loc[:, df_final.columns != 'evcs_count']
y = df_final['evcs_count']
x = sm.add_constant(x)
result = sm.OLS(y, x).fit()
print(result.summary())
```

```python
x = df_final.loc[:, df_final.columns != 'ev_registration_count']
y = df_final['ev_registration_count']
x = sm.add_constant(x)
result = sm.OLS(y, x).fit()
print(result.summary())
```

### LASSO REGRESSION

```python
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedKFold
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
data_train, data_val = train_test_split(df_final, test_size = 0.2, random_state = 2)

#Dependent Variable
Y_train = data_train.iloc[:, -1].values
#Independent Variables
X_train = data_train.iloc[:,0 : -1].values
#Independent Variables for Test Set
X_test = data_val.iloc[:,0 : -1].values

# Evaluating The Model With RMLSE
def score(y_pred, y_true):
  error = np.square(np.log10(y_pred +1) - np.log10(y_true +1)).mean() ** 0.5
  score = 1 - error
  return score

actual_cost = list(data_val['evcs_count'])
actual_cost = np.asarray(actual_cost)


#Initializing the Lasso Regressor with Normalization Factor as True
lasso_reg = Lasso()
# model = make_pipeline(StandardScaler(with_mean=False), Lasso())
#Fitting the Training data to the Lasso regressor
lasso_reg.fit(X_train,Y_train)
#Predicting for X_test
y_pred_lass =lasso_reg.predict(X_test)
#Printing the Score with RMLSE
print("\n\nLasso SCORE : ", score(y_pred_lass, actual_cost))
```
